```python
import os
import json
```

```python
root_folder = "E:\YandexDisk\datasets\kaggle"

def get_keywords_from_json_files(root_folder):
    all_keywords = []

    for root, dirs, files in os.walk(root_folder):
        for file in files:
            if file.endswith(".json"):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    try:
                        data = json.load(f)
                        if 'keywords' in data and isinstance(data['keywords'], list) and len(data['keywords']) == 5:
                            all_keywords.extend(data['keywords'])
                            print(data['title'], data['keywords'])
                    except json.JSONDecodeError as e:
                        print(f"Error reading {file_path}: {e}")

    return all_keywords

all_keywords = set(get_keywords_from_json_files(root_folder))

# Выводим или обрабатываем общий список ключевых слов
# for keyword in all_keywords:
#     print(keyword)
```
