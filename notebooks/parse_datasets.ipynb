```python
import kaggle
import os
import csv
```

```python
list_of_queries = []

with open('queries') as file:
    for line in file.readlines():
        list_of_queries.append(line.strip())
```

```python
!kaggle datasets metadata {f"hacker-news/hacker-news"} -p {f"E:\\YandexDisk\\datasets\\metadata-for-search\\hacker-news\\hacker-news"}
```

```python
# list_of_queries = ['regression', 'text classification', 'recommendation', 'Time Series Analysis', 'Natural Language Processing', 'Data analysis and visualization']

# 20 датасетов на странице

headers = ["query name", "rating_active", "path to metadata"]

with open(f"E:\\YandexDisk\\datasets\\metadata-for-search-info-active-rating.csv", 'w', newline='') as csv_file:
    
    writer = csv.writer(csv_file)
    writer.writerow(headers)
    
    
    for query in list_of_queries:
        path_query = f"E:\\YandexDisk\\datasets\\metadata-for-search"
        
        if not os.path.exists(path_query):
            os.makedirs(path_query)
            
        rating_count = 1
            
        for page in range(1, 2):
            list_of_datasets = !kaggle datasets list -s "{query}" -v --sort-by active -p {page}
            
            if list_of_datasets[0] == 'No datasets found':
                break
            
            for i in range(2, len(list_of_datasets), 2):
                dataset_information = list_of_datasets[i].split(',')
                print(dataset_information)
                remote_path_of_dataset = dataset_information[0]
                
                if "codec can't encode character" in remote_path_of_dataset:
                    continue
                
                local_path = f"{path_query}\\{remote_path_of_dataset}"
    
                if not os.path.exists(local_path):
                    os.makedirs(local_path)
                
                !kaggle datasets metadata {remote_path_of_dataset} -p {local_path}
                
                writer.writerow((query, rating_count, remote_path_of_dataset))

                rating_count += 1
            
                # !kaggle datasets download {remote_path_of_dataset} -p {local_path}
```
