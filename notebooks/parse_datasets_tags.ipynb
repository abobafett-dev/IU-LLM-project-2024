```python
import os

with open('../tags_en.txt', 'r') as file:
    keywords_list = [line.strip().lower() for line in file]

local_path = '../datasets/kaggle_datasets_parse_by_tags'

max_tags = 20

for keyword in keywords_list:
    counter = 0
    flag = True

    for page in range(1, max_tags):
        if flag:
            list_of_datasets = !kaggle datasets list --tags "{keyword}" -v --sort-by votes -p {page}

            if list_of_datasets[0] == 'No datasets found':
                flag = False
                break

            for i in range(2, len(list_of_datasets), 2):
                if flag:
                    dataset_information = list_of_datasets[i].split(',')
                    # print(dataset_information)
                    remote_path_of_dataset = dataset_information[0]

                    if "codec can't encode character" in remote_path_of_dataset:
                        continue

                    local_dataset_path = f"{local_path}\\{remote_path_of_dataset}"

                    if not os.path.exists(local_path):
                        os.makedirs(local_path)

                    response = !kaggle datasets metadata {remote_path_of_dataset} -p {local_dataset_path}
                    print(response[0])
                    if response != ['404 - Not Found - Dataset not found']:
                        counter += 1
                        if counter == max_tags:
                            flag = False
                else:
                    break
        else:
            print(f"Downloaded {counter} datasets with tag {keyword}")
            break
```
